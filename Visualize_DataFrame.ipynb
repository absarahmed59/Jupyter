```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,2*3.14,1000)
yc = list(np.cos(x))
ys = list(np.sin(x))

plt.plot(x,yc,'-')
plt.scatter(x,ys,color='black',s=0.1)
plt.show()
```
